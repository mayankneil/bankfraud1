```python
# 2.1 Cleaning The Data Set 1 : Application Data (application_data.csv)
```

```python
# columns related to the DAYS are negative. here we convert them into positive values.
appd.loc[:,"DAYS_BIRTH": "DAYS_ID_PUBLISH"] = abs(appd.loc[:,"DAYS_BIRTH": "DAYS_ID_PUBLISH"])
```

```python
# converting age in year format
appd['AGE'] = appd['DAYS_BIRTH'] // 365
```

```python
appd_miss = pd.DataFrame((appd.isnull().sum()/len(appd))*100).reset_index()
appd_miss.columns = ['Column Name', 'Null Value %age']
```

```python
round((appd.isnull().sum()/len(appd)*100.00), 5)
```

```python
appd_miss_30 = appd_miss[appd_miss['Null Value %age']>=30]
appd_miss_30
```

```python
len(appd_miss_30)
```

```python
appd_missing = pd.DataFrame((appd.isnull().sum()/len(appd))*100).reset_index()
appd_missing.columns = ['Column Name', 'Percentage of NULL values']
```

```python
# For Columns Having Null Values 
```

```python
appd_missing = appd_missing[appd_missing['Percentage of NULL values'] > 0]
appd_missing
```

```python
appd.AMT_ANNUITY.isnull().sum()
```

```python
#12 Rows have null values so we can drop the columns as it's won't really effect the later analysis.
```

```python
#Dropping the columns

appd = [~appd.AMT_ANNUITY.isnull()]
```

```python
appd.AMT_GOODS_PRICE.isnull().sum()
```

```python
plt.title('AMT_GOODS_PRICE')
ax = sns.boxplot(y = appd.AMT_GOODS_PRICE)
plt.show()
```

```python
# As we can see there are soo many outliers. So, imputing the NULL values with median
```

```python
appd.AMT_GOODS_PRICE = appd.AMT_GOODS_PRICE.fillna(appd.AMT_GOODS_PRICE.median())
```

```python
# NAME_TYPE_SUITE
appd.NAME_TYPE_SUITE.isnull().sum()
```

```python
# NAME_TYPE_SUITE is a categorical variable and have a lower NULL percentage.
# So, we will use mode to impute the NULL values for this variable

appd.NAME_TYPE_SUITE = appd.NAME_TYPE_SUITE.fillna(appd.NAME_TYPE_SUITE.mode()[0])
```

```python
# OCCUPATION_TYPE

# This column has more than 30% NULL values. So, we will make a separate category 'Unknown' for the NULL values

appd.OCCUPATION_TYPE = appd.OCCUPATION_TYPE.fillna('Unknown')
```

```python
# CNT_FAM_MEMBERS
appd.CNT_FAM_MEMBERS.isnull().sum()
```

```python
# Since just 2, we can delete this.
appd = appd[~appd.CNT_FAM_MEMBERS.isnull()]
```

```python
appd = appd[~appd.DAYS_LAST_PHONE_CHANGE.isnull()]
```

```python
# AMT variables
appd[['AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON',
         'AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR']].describe()
```

```python
# number of enquires and it can only be integers. So, imputing the NULL values with median as median is an integer

amt = ['AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK',
       'AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR']

for i in amt:
    appd[i].fillna(appd[i].median(),inplace = True)
```

```python
# SOCIAL variables
appd[['OBS_30_CNT_SOCIAL_CIRCLE', 'DEF_30_CNT_SOCIAL_CIRCLE',
     'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE']].describe()
```

```python
# these variables indicating the number of people and it can only be integers. So, imputing the NULL values with median.

soc = ['OBS_30_CNT_SOCIAL_CIRCLE', 'DEF_30_CNT_SOCIAL_CIRCLE',
     'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE']

for i in soc:
    appd[i].fillna(appd[i].median(),inplace = True)
```

```python
# Checking for the NULL values again to be sure!
print(appd.isnull().sum())
print('Shape: ', appd.shape)
```

```python
# Let's check the relationship of EXT_SOURCE variables with the TARGET variables
source = appd[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 'TARGET']]
```

```python
sns.heatmap(source.corr(),
            xticklabels= source.columns,
            yticklabels= source.columns,
            annot=True,
            cmap = 'Greens')
plt.show()
```

```python
# Adding these SOURCE variables to the above 50 variables which are to be dropped
appd_delete = appd_miss_30['Column Name'].to_list() + ['EXT_SOURCE_2', 'EXT_SOURCE_3']
len(appd_delete)
```

```python
# people who have submitted document 3 are more likely to not default on loan and hence, this must be an important document.
# while others are not as important as it is so we can drop them by grouping and adding them to the dropped columns set(appd_delete)

flag_cols = ['FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5', 'FLAG_DOCUMENT_6',
             'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8', 'FLAG_DOCUMENT_9', 'FLAG_DOCUMENT_10',
             'FLAG_DOCUMENT_11', 'FLAG_DOCUMENT_12','FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14',
             'FLAG_DOCUMENT_15', 'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17', 'FLAG_DOCUMENT_18',
             'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21']
```

```python
appd_delete = appd_delete + flag_cols
len(appd_delete)
```

```python
contact = appd.loc[:, 'FLAG_MOBIL':'FLAG_EMAIL']
contact['TARGET'] = appd['TARGET']
# Correlation of Contact-related information with the TARGET variable
plt.figure(figsize = (8,7))
sns.heatmap(contact.corr(),
            cmap="Reds",
            annot=True)
plt.show()
```

```python
# As correlation between target and phone related columns are not much we can drop them as well.
contact_cols = ['FLAG_MOBIL', 'FLAG_EMP_PHONE', 'FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE',
       'FLAG_PHONE', 'FLAG_EMAIL']
appd_delete = appd_delete + contact_cols
len(appd_delete)
```

```python
#Dropping the unwanted columnms now
appd.drop(labels=appd_delete, axis=1, inplace=True)
```

```python
# Making AGE GROUPS for further Analysis
bins = [0, 20, 30, 40, 50, 100]
labels = ['0-20', '20-30', '30-40', '40-50', '50+']
appd['AGE_GROUP'] = pd.cut(appd['AGE'], bins = bins, labels= labels)
appd.AGE_GROUP.value_counts(normalize=True)*100
```
